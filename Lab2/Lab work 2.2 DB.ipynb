### Завдання 6 

Напишіть функції, що:
* рохраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```python
import random
import math

def minmax(l):
    x = min(l)
    y = max(l)
    return x,y

def average(l):
    add = 0
    for i in l:
        add = add + i
    mean = add / len(l)
    return mean

def median(l):
    if len(l) % 2 == 0:
        median1 = l[len(l)//2]
        median2 = l[len(l)//2 - 1]
        median = (median1 + median2) / 2
    else:
        median = l[len(l)//2]
    return median

def frequency(l):
    counts = dict()
    for number in l:
        if number in counts:
            counts[number] += 1
        else:
            counts[number] = 1
    return counts

def standart_deviation(l):
    x = average(l)
    add = 0
    for i in l:
        dif = (l[i] - x)**2
        add = add + dif
    std = (add / len(l))**0.5
    return std


#генеруэмо тисячу чисел в діапазоні від 0 до 10
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]
print(list_with_random_numbers)

x = minmax(list_with_random_numbers)

y = average(list_with_random_numbers)

z = median(list_with_random_numbers)

a = frequency(list_with_random_numbers)

b = standart_deviation(list_with_random_numbers)

print(x,'minmax')
print(y,'average')
print(z,'median')
print(a,'frequency')
print(b, 'std')
                    
```
