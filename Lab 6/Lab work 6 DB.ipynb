### Завдання:

```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import seaborn as sns
```

```python
np.random.seed(0)

df = pd.DataFrame(data={'a':np.random.randint(0, 100, 30),
                        'b':np.random.randint(0, 100, 30),
                        'c':np.random.randint(0, 100, 30)})
df.head()
```

Створити візуалізацію, аналогічно рисунку 
 - перші 3 графіки візуалізувати, викорстовуючи значення із df
 - останній це пряма пропорційність

```python
fig, (axs) = plt.subplots(2, 2, figsize = (12,8), constrained_layout=True)

axs[0,0].plot(df['a'])

axs[0,1].plot(df['b'])

axs[1,0].plot(df['c'])

xaxis = np.array([0,30])
yaxis = np.array([0,30])
axs[1,1].plot(xaxis,yaxis)
plt.show()
```

Створити візуалізацію, аналогічно рисунку 

```python
plt.figure(figsize = (15,10))
plt.plot(df['a'])
plt.show()
```

Створити візуалізацію, аналогічно рисунку 
- використовуйте колонки `a` та `b`

```python
fig, (ax1,ax2) = plt.subplots(2, 1, figsize = (12,6), constrained_layout=True)

ax1.plot(df['a'])

ax2.plot(df['b'])

plt.show()
```

Створити візуалізацію, аналогічно рисунку

```python
plt.figure(figsize = (15,10))
plt.plot(df['a'])
plt.plot(df['b'])

plt.show()
```

Створити візуалізацію, аналогічно рисунку 
- використовуйте колонки `a` та `b`
- задайте стиль 'darkgrid' за допомогою команди `sns.set_style`

```python
plt.figure(figsize = (15,10))
plt.plot(df['a'])
plt.plot(df['b'])
sns.set_style(style='darkgrid')
plt.show()
```

Створити візуалізацію, аналогічно рисунку 
- для колонки `a` використайте червоний колір та лінію формату `-.`

- для колонки `b` використайте помаранчевий колір та товщину `10`

- для колонки `c` використайте жовтий колір та товщину `1` і маркер `o`


```python
plt.figure(figsize = (15,10))
plt.plot(df['a'],color="red",linestyle='-.')
plt.plot(df['b'],color="orange",linewidth=10)
plt.plot(df['c'],color="yellow",linewidth=1,marker='o')
plt.show()
```

Створити візуалізацію, аналогічно рисунку 
- і не забудьте про легенду :)

```python
fig, (ax1,ax2,ax3) = plt.subplots(3, 1, figsize = (12,8), constrained_layout=True)

ax1.plot(df['a'],label= "line A")

ax2.plot(df['b'],label="line B")

ax3.plot(df['c'])
ax1.legend(loc="lower right")
ax2.legend(loc="center left")
plt.show()
```

Створити візуалізацію, аналогічно рисунку 
- використайте томатний колір та відстань між стовбцями 0.5

```python
plt.figure(figsize = (15,10))
# df['a'].hist()
plt.bar(range(len(df["a"])),df["a"],width=0.5,color=("tomato"),label="a")
plt.legend(loc="upper right")

plt.show()
```
